## Filtering operations on the dataframe

```python
# checking for countries with ranks less than 10 

df[df['Rank']<=10]
```

```python
#filtering for countries within a specific list, kind of like "where X-field in ()" in SQL

specific_countries = ["Bangladesh","Brazil"]
df[df["Country"].isin(specific_countries)]
```

```python
#Filtering for countries that contain specific letters, kind of similar to using "Like" in SQL 

df[df["Country"].str.contains("United")]
```

```python
#Filtering on indexes to only show specific countries's fields 

df2 = df.set_index('Country')
df2.filter(items =['Continent', 'CCA3','Capital'],axis=1)
```

```python
#Filtering on indexes to only show specific countries

df2.filter(items =['Zambia', 'Andorra'],axis=0)
```

```python
#Filtering on indexes to only show specific countries which have similar text 


df2.filter(like ='am',axis=0)
```

## Group By and Aggregating

```python
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
continents = df.groupby('Continent')
```

```python
# group by is used to make aggregations 
#mean only takes in numerical fields 

continents.mean()
```

```python
average_population = df.groupby('Continent').mean()
average_population
```

```python
min_population = df.groupby('Continent').min()
min_population
```

```python
#you can choose to run several aggregations on a single or more column using the aggregate function and a dictionary

df.groupby('Continent').agg({'2022 Population': ['mean','max','min','count','sum'],'2020 Population': ['mean','max','min','count','sum']})
```

```python
df.groupby(['Continent','2022 Population']).min()
```
